## Données prix vente imobilier

```python
data=pd.read_csv("./train_imo.csv")
```

Les données contiennent des informations sur des pappartement vendu. L'objectif est de mettre en place un réseau de neuronnes capable de prédire le prix de ventes d'un appartement à partir de ces variables. La quantifiaction de l'erreur de prédiction devra se faire à l'aide de l'erreur quadratique moyenne mais sur le logarithme du prix.

```python
N=data.shape[0]
data.shape
```

La base de données est constitué de 1460 biens immobiliers décrit par 81 varibales. L'une de ces variable SalePrice est la variable cible à prédire à partir des autres.

```python
data.describe()
```

Parmis les 80 descripeturs utilisables 37 correspondent à des variables continues. Les autres variables peuvent être considérées comme des variables catégorielle est encoder sous forme de vecteur binaire (onehot encoding). Le code suivant permet de construire une matrice de données mise en forme pour l'apprentissage' 

```python
from sklearn.preprocessing import OneHotEncoder,LabelEncoder,scale
oh = []
ohencoders = []
labencoders= []
Xnum = []
for v in range(1,80):
    if data.dtypes[v]=="object" :
        labenc = LabelEncoder()
        data.iloc[:,v].fillna("unknown",inplace=True)
        labenc.fit(data.iloc[:,v])
        Xl = labenc.transform(data.iloc[:,v]).reshape(-1, 1)
        ohenc = OneHotEncoder(sparse=False)
        ohenc.fit(Xl)
        Xt=ohenc.transform(Xl)
        labencoders.append(labenc)
        ohencoders.append(ohenc)
        oh.append(Xt)
    else:
        Xnum.append(scale(data.iloc[:,v].fillna(0).apply(float).values).reshape(N,1))
```

```python
import numpy as np
Xnum = np.concatenate(Xnum,axis=1)
Xoh  = np.concatenate(oh,axis=1)
```

```python
X = np.concatenate([Xnum,Xoh],axis=1)
```
